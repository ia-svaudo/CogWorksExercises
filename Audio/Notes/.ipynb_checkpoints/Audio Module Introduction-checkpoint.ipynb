Sound wave is a pressure wave.

Speed of sound: 343m/s

p(x) = Asin(2* pi* f* t + phi)

f = frequency

t = time

A = amplitude

p = pressure

f = 1/T

phi = phase shift

sin(theta) = cos(theta-pi/2)



```python
c_4 = np.array([1,2,3,4,5])
Audio(c_4, rate=4000)
```

```python
%matplotlib inline
times = np.linspace(0, 12, 1000)
freq = 1 / 3  # Hertz
amp = 0.02  # Pascals
pressures = amp * np.sin(2 * np.pi * freq * times)  # Pascals; shape-(1000,)


fig, ax = plt.subplots()

ax.plot(times, pressures)
ax.set_xlabel("Time [seconds]")
ax.set_ylabel("Pressure Relative to Standing Air [Pascals]")
ax.set_title("Temporal Waveform of Pure Tone")
ax.grid(True)
```

Superposition is adding functions together to get a new wave
